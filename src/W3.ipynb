```python
def OptimalPorfolio(Mu_expected,Cov_R):
    InC = inv(Cov_R)
    one = np.ones((len(Mu_expected),1))
    One_r = one.reshape(1,-1)
    
    A = np.dot(np.dot(One_r,InC),Mu_expected)
    B = np.dot(np.dot(Mu_expected.reshape(1,-1),InC),Mu_expected)
    D = B*np.dot(np.dot(One_r,InC),one) - A**2
    
    gmin = (B * np.dot(InC, one) - A* np.dot(InC, Mu_expected))/D
    hmin = (np.dot(np.dot(One_r,InC),one) * np.dot(InC, Mu_expected) - A* np.dot(InC, one))/D
    
    F1= np.dot(np.dot(gmin.reshape(1,-1),Cov_R),gmin)
    F2= (np.dot(np.dot(hmin.reshape(1,-1),Cov_R),gmin)**2)/np.dot(np.dot(hmin.reshape(1,-1),Cov_R),hmin)
    min_var = F1 - F2
    
    return gmin,hmin,min_var
```

```python
def effcientfrontier(g,h,R_cov,mini,maxa):
    sd = []
    mu = []
    weight = []
    C = 0.00000001
    yr= 252
    previous = np.inf
    for i in range(mini,maxa):
        w = g + h*C*i
        sd_new = yr*np.dot(np.dot(w.reshape(1,-1),R_cov),w)[0][0]
        if sd_new > previous:
            sd0=np.sqrt((np.exp(sd_new)-1)*np.exp(2*C*i*yr + sd_new))*100
            sd.append(sd0)
            weight.append(w)
            mu_new=(np.exp(C*i*yr + sd_new/2)-1)*100
            mu.append(mu_new)
        previous = sd_new
        
    return sd,mu,weight
```

```python
g4,h4,min_sigma4 = OptimalPorfolio(np.array(Return.mean()).reshape(-1,1),np.array(Return.cov()))
sigma_4,mu_4,weight_4 = effcientfrontier(g4,h4,np.array(Return.cov()),2000,200000)
```

```python
risk_free = 0.01875 

def tangency_ratio(sigma,mu,rf):
    ratio = 0
    ind = 0
    for i in range(len(sigma)):
        ratio_update = (mu[i] - rf*100)/sigma[i]
        if ratio_update > ratio:
            ratio = ratio_update
            ind = i           
    return ratio,ind

ratio,ind = tangency_ratio(sigma_4,mu_4,risk_free)
targent_v = (ratio * np.linspace(6,30,2000).reshape(-1,1) + risk_free*100).reshape(-1,1)

plt.plot(np.linspace(6,30,2000).reshape(-1,1),targent_v,label = "Capital Market Line",color='orange')
plt.plot(sigma_4,mu_4,label = "Efficient Froniter",color='black')

tangent_sigma=sigma_4[ind]
tangent_mu=mu_4[ind]
plt.scatter(tangent_sigma, tangent_mu, color='r',marker = "x")
plt.text(tangent_sigma, tangent_mu,'Tangency portfolio')
plt.ylabel("Expected return")
plt.xlabel("Risk Tolerance")
plt.title("1 year + 1 week historical data EffcientFrontier")
plt.legend()
plt.show()
```
