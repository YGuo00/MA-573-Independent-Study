```python
# Goal: Provide a proof that the finite difference for the second order has convergence order 1.
import numpy as np
import matplotlib.pyplot as plt         
```

```python
def sfd(f, x, h):
    return (f(x + h) - 2 * f(x) + f(x - h))/(h**2)      
    pass
```

```python
h = .5

x_cod = np.linspace(0, 2*np.pi, 100)
y_cod = np.cos(x_cod)
plt.plot(x_cod, -np.sin(x_cod), label='$-sin(x)$');
plt.plot(x_cod, sfd(np.sin, x_cod, h), label = 'sfd');
plt.legend();
```

```python
err = sfd(np.sin, x_cod, h) - y_cod
yy = np.log2(np.abs(err))
plt.loglog(x_cod, yy)
```

```python
import scipy.stats as ss
out = ss.linregress(x_cod, yy)
print('the convergence order is ' + str(-out[0]))
```
