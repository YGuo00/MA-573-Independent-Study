```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def exactbm1d(T, N):
    mesh_pi = np.linspace(0, T, N + 1)   # np.linspace: interval(start pt., end pt., numbers in interval divided evenly)
    mesh_size = T / N
    BM_path = np.zeros(N + 1)   # This is W
    for i in range(N):
        BM_path[i + 1] = BM_path[i] + np.sqrt(mesh_size) * np.random.normal()
    return mesh_pi, BM_path
```

```python
h = 0.1
T = 110
N = int(T / h)

for i in range(10):
    t_array, w_array = exactbm1d(T, N)
    s = 100
    starting_index = int(s / h) + 1

    t_array = t_array[starting_index:]
    w_array = w_array[starting_index:]
    y_array = w_array / np.sqrt(2 * t_array * np.log(np.log(t_array)))
    # 

    plt.plot(t_array, y_array)
```
